# Python Essentials

## Data Types

Programs need to distinguish between data types since

- they are stored in memory differently
- different types are compatible with different operations

Like most other programming languages, Python has:

* floating point numbers, for example `1.5`
* integers, for example `1`
* strings, for example `"Hello"`
* lists, for example `[1.5, 'a', 'Hello']` etc.

As you can see, Python even supports mixed-type lists.

### Primitive Data Types

One simple data type is **Boolean values**, which can be either `True` or `False`

```python
x = True
x
```

We can check the type of any object in memory using the `type()` function.

```python
x = True
type(x)
```

In the next line of code, the interpreter evaluates the expression on the right of = and binds y to this value

```python
y = 100 < 10
y
```

```python
type(y)
```

### Containers

Python has several basic types for storing collections of (possibly heterogeneous) data.

A related data type is **tuples**, which are "immutable" lists

```python
x = ('a', 'b')  # Parentheses instead of the square brackets
x = 'a', 'b'    # Or no brackets --- the meaning is identical
x
```

```python
type(x)
```

In Python, an object is called **immutable** if, once created, the object cannot be changed.

Conversely, an object is **mutable** if it can still be altered after creation.

Python lists are mutable

```python
x = [1, 2]
x[0] = 10
x
```

But tuples are not

```python
x = (1, 2)
try: 
    x[0] = 10    
except TypeError:
    print("TypeError: 'tuple' object does not support item assignment")
```

We'll say more about the role of mutable and immutable data a bit later.

Tuples (and lists) can be "unpacked" as follows

```python
integers = (10, 20, 30)
x, y, z = integers
x
```

```python
y
```

## Iterating

One of the most important tasks in computing is stepping through a
sequence of data and performing a given action.

One of Python's strengths is its simple, flexible interface to this kind of iteration via
the `for` loop.

### Looping over Different Objects

Many Python objects are "iterable", in the sense that they can be looped over.
For example,

```python
option_types = ['European', 'American', 'Asian', 'Bermudan']
for kind in option_types:    
    print(kind)
```

### Looping without Indices

One thing you might have noticed is that Python tends to favor looping without explicit indexing.

For example,

```python
x_values = [1, 2, 3]  # Some iterable x
for x in x_values:
    print(x * x)
```

is preferred to

```python
for i in range(len(x_values)):
    print(x_values[i] * x_values[i])
```

Python provides some facilities to simplify looping without indices.

One is `zip()`, which is used for stepping through pairs from two sequences.

For example, try running the following code

```python
countries = ('Japan', 'Korea', 'China')
cities = ('Tokyo', 'Seoul', 'Beijing')
for country, city in zip(countries, cities):
    print(f'The capital of {country} is {city}')
```

If we actually need the index from a list, one option is to use `enumerate()`.

To understand what `enumerate()` does, consider the following example

```python
letter_list = ['a', 'b', 'c']
for index, letter in enumerate(letter_list):
    print(f"letter_list[{index}] = '{letter}'")
```

### List Comprehensions

We can also simplify the code for generating the list of random draws considerably by using something called a *list comprehension*, for example

```python
doubles = [2 * x for x in range(8)]
doubles
```

## Sampling normals and statistics

Let's start with sampling the standard normal variables required for the 
increments of a single Brownian motion path:

```python
n_times = 500
e = []

for i in range(n_times):
    r = np.random.randn()
    e.append(r)

plt.plot(e)
plt.show()
```

We use the statement `e = []` to create an empty list.

In the loop, we add values to a list using the syntax `list_name.append(some_value)`

Here `append()` is what's called a *method*, which is a function "attached to" an object---in this case, the list `x`. We'll learn all about methods later on when we study classes.

Lists are zero-based, so the first element is referenced by `x[0]`

```python
e[0]   # first element of x
```

```python
e[1]   # second element of x
```

### The For Loop

Now let's consider the `for` loop, which was

```python
for i in range(n_times):
    r = np.random.randn()
    e.append(r)
```

Python executes the two indented lines `n_times` times before moving on.

These two lines are called a `code block`, since they comprise the "block" of code that we are looping over.

Unlike most other languages, Python knows the extent of the code block *only from indentation*.

In our program, indentation decreases after line `e.append(e)`, telling Python that this line marks the lower limit of the code block.

### A Comment on Indentation

In discussing the `for` loop, we explained that the code blocks being looped over are delimited by indentation.

In fact, in Python, **all** code blocks (i.e., those occurring inside loops, if clauses, function definitions, etc.) are delimited by indentation.

Thus, unlike most other languages, whitespace in Python code affects the output of the program.

Once you get used to it, this is a good thing: It

* forces clean, consistent indentation, improving readability
* removes clutter, such as the brackets or end statements used in other languages

On the other hand, it takes a bit of care to get right, so please remember:

* The line before the start of a code block always ends in a colon
  * `for i in range(10):`
  * `if x > y:`
  * `while x < 100:`
  * etc., etc.
* All lines in a code block **must have the same amount of indentation**.
* The Python standard is 4 spaces, and that's what you should use.


Python lists are very handy data structures, but are not what is normally used for numerical

## Simulate the Geometric Brownian Motion Process

Suppose we want to simulate and plot the geometric brownian motion process
$$ dS_t = S_t \left[ r dt + \sigma dW_t \right]$$
on a discrete uniform time-step schedule $t_0, t_1, \ldots, t_n$ such that $t_i - t_{i-1} = \delta t, \forall i$.
$$ S_{t_i} = S_{t_{i-1}} \exp\left( (r-\frac{\sigma^2}{2})(t_i - t_{i-1}) + \sigma \sqrt{t_i - t_{i-1}} \epsilon_i \right),$$
where $\epsilon_i$, for $i \in [1, n]$  are independent standard normal variables.

We explore multiple ways of coding this in Python, simulating different paths and plotting them.

Let's fix some parameter values for our experiments:

```python
S_0 = 100
T = 10
vol = 0.2
r = 0.01
```

Start by first simulating just the terminal value

```python
S = S_0 * math.exp((r - vol**2 / 2)*T + vol*math.sqrt(T)*np.random.randn())
S
```

To do multiple paths, we just need to repeat that once for each path. So let's do that

```python
n = 20
St = np.empty(n)
for i in range(n):
    St[i] = S_0 * math.exp((r - vol**2 / 2)*T + vol*math.sqrt(T)*np.random.randn())

print('St shape:', St.shape)

S0 = np.full_like(St ,S_0)
print('S0 shape:', S0.shape)

S = np.stack([S0, St])
print('S shape:', S.shape)

plt.plot([0, 10], S)
plt.show()
```

We can also do the same without a loop using 'vectorized' numpy syntax

```python
St = S_0 * np.exp((r - vol**2 / 2)*T + vol*math.sqrt(T)*np.random.randn(n))
print('St shape:', St.shape)

S0 = np.full_like(St ,S_0)
print('S0 shape:', S0.shape)

S = np.stack([S0, St])
print('S shape:', S.shape)

plt.plot([0, 10], S)
plt.show()
```

Let's move on to simulating paths with multiple time steps.


We start by precomputing some terms

```python
n_dt = 520
dt = T/n_dt
vol_sqrt_dt = vol * math.sqrt(dt)
ito_term = (r - vol**2 / 2) * dt
```

Here are a few lines of code that perform the task for a single path

```python
e = np.random.randn(n_dt)
S = np.zeros_like(e)
S[0] = S_0

for t in range(1, n_dt):
    S[t] = S[t-1] * math.exp(ito_term + vol_sqrt_dt * e[t])

fig, (ax1, ax2) = plt.subplots(2,1)
ax1.plot(e)
ax2.plot(S)
plt.show()
```

And now, let's do multiple paths. 

We first sample all the normal variates we need for all paths and time steps

```python
n_paths = 100
e = np.random.randn(n_dt, n_paths)
e.shape
```

The same calculation we did for a single path and time, we can now conduct in a double loop 
to all time steps for each path

```python
S = np.zeros_like(e)
S[0,:] = S_0 

for path in range(0, n_paths):
  for t in range(1, n_dt):
    S[t, path] = S[t-1, path] * math.exp(ito_term + vol_sqrt_dt * e[t, path])

plt.plot(S)
plt.show()
```

We can also permute the loops to do all paths for each time step

```python
S = np.zeros_like(e)
S[0,:] = S_0 

for t in range(1, n_dt):
    for path in range(0, n_paths):
        S[t, path] = S[t-1, path] * math.exp(ito_term + vol_sqrt_dt * e[t, path])

plt.plot(S)
plt.show()
```

Another way is also to iterate forward in the time dimension until we reach T:

```python
S = np.zeros_like(e)
S[0,:] = S_0 
t = dt
i = 1

while t < T:
    for path in range(0, n_paths):
        S[i, path] = S[i-1, path] * math.exp(ito_term + vol_sqrt_dt * e[i, path])
    t = t + dt
    i = i + 1    

plt.plot(S)
plt.show()
```

### While Loops

Note that

* the code block for the `while` loop is again delimited only by indentation
* the statement  `i = i + 1` can be replaced by `i += 1`


We saw earlier how 'vectorized' calculations can be used to do all paths in one 
expressions. Let's apply that to every time step:

```python
S = np.zeros_like(e)
S[0,:] = S_0 

for t in range(1, n_dt):
    S[t, :] = S[t-1, :] * np.exp(ito_term + vol_sqrt_dt * e[t, :])

plt.plot(S)
plt.show()
```

We can also write it, with implicit slicing

```python
S = np.zeros_like(e)
S[0] = S_0 

for t in range(1, n_dt):
    S[t] = S[t-1] * np.exp(ito_term + vol_sqrt_dt * e[t])

plt.plot(S)
plt.show()
```

To also vectorize along the time direction, we need to transform the problem since
the above is formulated in terms of incremental calculations:

```python
ti = np.linspace(0,T, n_dt).reshape(-1,1)
dW_t = math.sqrt(dt) * e 
W_t = np.cumsum(dW_t, axis=0)

S = S_0 * np.exp((r-0.5*vol**2)*ti + vol * W_t) 

plt.plot(S)
plt.show()
```

### Solution 1

- Use a 'Kolmogorov-Smirnoff' test to statistically verify that the sample 
  is indeed lognormal with the correct parameters

```python
import math
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = (14,10)
```

Here we sample all the terminal values

```python
St = S_0 * np.random.lognormal( (r - 0.5* vol**2) * T, vol*math.sqrt(T), N)
```

Let's plot a few paths

```python
St_sub = St[:100]
S0 = np.full_like(St_sub ,S_0)
S = np.stack([S0, St_sub])

plt.plot([0, T], S)
plt.show()
```

and compare with theoretical values to make sure our simulation is right

```python
print('theoretical mean:', S_0 * math.exp(r*T))
print('theoretical variance:', S_0**2 * math.exp(2*r*T) * (math.exp(vol**2 * T)-1) )
```

Take a look at the qq-plot against the lognormal distribution

Compare density plots

```python
Y = St/S_0
shape, loc, scale = stats.lognorm.fit(Y, floc=0)
plt.hist(Y, 1000, density=True)
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = stats.lognorm.pdf(x, shape, loc, scale)
plt.plot(x, p)
plt.show()
```

KS-test:

```python
from scipy.stats import kstest
statistic,pvalue = kstest(Y, 'lognorm', args=(shape, 0, scale))
print('statistic:', statistic, 'p-value:', pvalue )
if pvalue < 0.05:
    print("p-value: {:.2f}, data does not look lognormal (reject H0 at 5% significance level)".format(pvalue))
else:
    print("p-value: {:.2f}, data looks lognormal (fail to reject H0 at 5% significance level)".format(pvalue))
```

### Solution 2

```python
from scipy import stats
T = 1
S_0 = 100
vol = 0.25
r = 0.02
N = np.array([1024 * 2**n for n in range(10)])
s_bar = np.empty(N.size)
s_u = np.empty(N.size)
s_l = np.empty(N.size) 

for k,n in enumerate(N):
    St = S_0 * np.random.lognormal( (r - 0.5* vol**2) * T, vol*math.sqrt(T), n)
    mean, var, std = stats.bayes_mvs(St, 0.95)    
    s_bar[k] = mean.statistic
    s_u[k] = mean.minmax[1]
    s_l[k] = mean.minmax[0]

fig, (ax1,ax2) = plt.subplots(2,1)

ax1.semilogx(N, s_bar)
ax1.fill_between(N, s_u, s_l, color='b', alpha=.1)

ax2.semilogx(N, s_u-s_l)
ax2.semilogx(N, 100/np.sqrt(N), color='r', linestyle='--')
plt.show()    
```

### Solution 3

```python
def bs_call(S, K, T, r, sigma):
    
    d1 = (np.log(S/K) + (r + (sigma**2)/2) * T) / (sigma*np.sqrt(T))
    d2 = d1 - sigma*np.sqrt(T)
    call = S * stats.norm.cdf(d1) - K * np.exp(-r*T) * stats.norm.cdf(d2)
    
    return call
```

Given the parameters

```python
T = 1
S_0 = 100
vol = 0.2
r = 0.02
K = 100
N = np.array([1024 * 2**n for n in range(10)])
C_hat = np.empty(N.size)
C_l = np.empty(N.size)
C_u = np.empty(N.size)

bs = bs_call(S_0, K, T, r, vol)
bench = np.full(N.size, bs)

for k,n in enumerate(N):
    St = S_0 * np.random.lognormal((r - 0.5* vol**2) * T, vol*np.sqrt(T), n)
    Ct = math.exp(-r*T) * np.maximum(St - K, 0) 
    mean, var, _ = stats.bayes_mvs(Ct, 0.99)
    C_hat[k] = mean.statistic
    C_l[k] = mean.minmax[0]
    C_u[k] = mean.minmax[1]

plt.semilogx(N,C_hat) 
plt.semilogx(N,bench,color='r', linestyle='--') 
plt.fill_between(N,C_l, C_u, color='b', alpha=.1)        
plt.show()
```

### Solution 4

An Arithmetic Asian call option has for underlying an arithmetic average of the underlying at different times.
Price by Monte-Carlo an Asian call on the monthly average, given the parameters

```python
from scipy import stats
import numpy as np
import math

T = 5
S_0 = 100
vol = 0.2
r = 0.02
n_dt = T*12
dt = 1/12
N = 100000
K = 100

ti = np.linspace(0,T, n_dt)
e = np.random.normal(size=(N,n_dt))
dW_t = math.sqrt(dt) * e
W_t = np.cumsum(dW_t, axis=1)

S = S_0 * np.exp((r-0.5*vol**2)*ti + vol * W_t)
A = np.mean(S, axis=1)
Ct = math.exp(-r*T) * np.maximum(A - K, 0) 
C =  math.exp(-r*T) * np.maximum(S[:,-1] - K, 0) 
mean, var, _ = stats.bayes_mvs(Ct, 0.99)
print('Asian Call by MC', mean.statistic, '(', mean.minmax[0], mean.minmax[1], ')')
mean, var, _ = stats.bayes_mvs(C, 0.99)
print('European Call by MC', mean.statistic, '(', mean.minmax[0], mean.minmax[1], ')')
bs = bs_call(S_0, K, T, r, vol)
print('European Call with BS formula', bs)
```
